## Выводы
1. **Точность системы:**  
   - Для шрифта 54pt: **100.00%** (идеальное распознавание).  
   - Для шрифта 20pt: **37.50%** (критическое падение качества).  

2. **Основные ошибки:**  
   - Путаница между символами: **I↔D**, **V↔Y**, **O↔D**, **B↔M**.  
   - Некорректная сегментация мелких символов (20pt).  

3. **Причины проблем:**  
   - Низкое разрешение изображения для мелкого шрифта.  
   - Перекрытие символов из-за уменьшения межсимвольного расстояния.
