# Лабораторная работа №9. Анализ шума

---

## Цель работы
1. Запись аудиосигнала музыкального инструмента (гитара).
2. Построение спектрограммы с использованием оконного преобразования Фурье.
3. Фильтрация шума и сравнение спектрограмм до и после обработки.
4. Определение моментов времени с максимальной энергией в частотной полосе 40–50 Гц.

---

## Результаты

### 1. Запись аудиосигнала
- **Формат файла:** `.wav` (моно).
- **Частота дискретизации:** 44100 Гц.
- **Длительность сигнала:** 18 сек.

---

### 2. Спектрограмма исходного сигнала
- **Метод:** Оконное преобразование Фурье с окном Ханна.
- **Параметры:** 
  - Размер окна (n_fft): 2048.
  - Шаг между окнами (hop_length): 512.
- **Визуализация:** Логарифмическая шкала частот.

- **Параметры фильтра:** 
  - Частота среза: 3300 Гц.

  - Высокочастотные шумы (выше 3300 Гц) подавлены.

### 4. Анализ энергии в полосе 40–50 Гц
- **Параметры анализа:**  
  - Шаг по времени: Δt = 0.1 сек.
  - Полоса частот: Δf = 40–50 Гц.
- **Результаты:**  
  - **Моменты с максимальной энергией:** 6.1 сек, 6.8 сек, 16.1 сек.  
  - Эти моменты соответствуют ударам по струнам гитары.
