## Цветовые модели и передискретизация изображений  

### Исходные данные  
- Исходное изображение: `input.bmp` (24-битное BMP)  
- Все операции выполнены без использования библиотечных функций передискретизации

## 1. Работа с цветовыми моделями  

### 1.1 Выделение компонентов R, G, B  
**Операция**: Извлечение цветовых каналов из RGB-изображения  
**Алгоритм**:  
- Обнуление двух из трёх каналов для выделения целевого цвета  

**Результаты**:  
| Исходное изображение | Красная компонента (R) | Зелёная компонента (G) | Синяя компонента (B) |  
|----------------------|------------------------|------------------------|----------------------|  
| ![Original](input.bmp) | ![R](r_component.bmp) | ![G](g_component.bmp) | ![B](b_component.bmp) |  

---

### 1.2 Приведение изображения к цветовой модели HSI  
**Операция**: Преобразование RGB → HSI и выделение яркости (I)  
**Формулы**:  
- Яркость: `I = (R + G + B) / 3`  
- Насыщенность: `S = 1 - 3·min(R,G,B)/(R+G+B)`  
    

**Результаты**:  
| Исходное изображение |  Яркостная компонента (I) |  
|----------------------|---------------------------|  
| ![Original](input.bmp) |  ![I](i_component.bmp) |  

---

### 1.3 Инверсия яркостной компоненты  
**Операция**: Инверсия значения яркости по формуле `I' = 1 - I`  

**Результаты**:  
| Исходное изображение | Инвертированная яркость |  
|----------------------|--------------------------|  
| ![Original](input.bmp) | ![Inverted](inverted.bmp) |  

---

## 2. Передискретизация  

### 2.1 Растяжение изображения в 2 раза (M=2)  
**Метод**: Билинейная интерполяция  
**Алгоритм**:  
1. Вычисление новых координат  
2. Взвешенное усреднение четырёх ближайших пикселей  

**Результаты**:  
| Исходное изображение (1900x1900) | Растянутое изображение (3800x3800) |  
|--------------------------------|-----------------------------------|  
| ![Original](input.bmp) | ![Scaled Up](scaled_up.bmp) |  

---

### 2.2 Сжатие изображения в 2 раза (N=2)  
**Метод**: Децимация (прореживание пикселей)  
**Алгоритм**: Выбор каждого N-го пикселя  

**Результаты**:  
| Исходное изображение (1900x1900) | Сжатое изображение (950x950) |  
|--------------------------------|-------------------------------|  
| ![Original](input.bmp) | ![Scaled Down](scaled_down.bmp) |  

---

### 2.3 Двухпроходная передискретизация (K=1.5)  
**Этапы**:

2. Сжатие в 2 раза (`decimate(img, 2)`)
