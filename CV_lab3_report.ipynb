# Лабораторная работа №3  
**Тема:** Фильтрация изображений и морфологические операции  
**Вариант:** 3 (Логический фильтр, окно 3×3)  

---

**Логический фильтр AND (3×3):**  
- Анализирует окрестность 3×3 вокруг каждого пикселя.  
- Выходной пиксель становится белым (255), только если **все** пиксели в окне белые.  
- Эффект: удаление шумов, устранение мелких деталей, сглаживание границ.  

**Разностное изображение:**  
- Формируется операцией XOR между исходным и отфильтрованным изображениями.  
- Выделяет зоны, где фильтр изменил яркость пикселей.

## Выводы  
1. **Эффективность фильтра AND:**  
   - Удаляет шумы и мелкие детали, но искажает границы объектов.  
   - Наиболее полезен для изображений с выраженным фоном и крупными элементами.  

2. **Разностное изображение:**  
   - Позволяет анализировать работу фильтра на пиксельном уровне.  
   - Выделяет как устраненные шумы, так и потерю деталей.  

3. **Ограничения:**  
   - Требует точной настройки порога бинаризации.  
   - Не подходит для обработки полутоновых изображений без предварительной бинаризации.
