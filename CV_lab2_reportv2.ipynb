## Задание  
1. **Приведение полноцветного изображения к полутоновому.**  
   - Новое изображение создается в режиме полутона (1 яркостный канал, формат BMP).  
   - Яркость каждого пикселя вычисляется взвешенным усреднением каналов исходного полноцветного изображения по формуле:
     
     $$ Y = 0.299 \cdot R + 0.587 \cdot G + 0.114 \cdot B $$
     
     где $R, G, B$ – значения красного, зеленого и синего каналов соответственно, $ Y $ – итоговая яркость пикселя.

2. **Приведение полутонового изображения к монохромному методом пороговой обработки.**  
   - Бинаризация выполняется с использованием **адаптивного монохромного преобразования с усреднением по окну 5×5**.
   - Пороговое значение $ T(x,y) $ для каждого пикселя вычисляется как среднее значение яркости в окрестности $5 \times 5$:
     
     $$ T(x,y) = \frac{1}{25} \sum_{i=-2}^{2} \sum_{j=-2}^{2} Y(x+i, y+j) $$
     
   - Бинаризация происходит по правилу:
     
     $$ B(x, y) = \begin{cases}  
     255, & Y(x, y) \geq T(x, y)  \\  
     0, & Y(x, y) < T(x, y)  
     \end{cases} $$
