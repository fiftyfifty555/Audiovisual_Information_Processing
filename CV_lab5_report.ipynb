# Лабораторная работа №5. Выделение признаков символов

**Вариант:** 23 (английские заглавные буквы)

---

## Выполнение

### 1. Генерация изображений символов
Сгенерированы изображения символов A-Z с использованием шрифта Times New Roman (размер 52). Каждое изображение обрезано по границам символа и сохранено в папку `symbols`.

### 2. Расчет признаков
Для каждого символа рассчитаны:
- Вес и удельный вес четвертей изображения.
- Координаты центра тяжести.
- Моменты инерции.
- Профили по осям X и Y.

### 3. Визуализация
Профили сохранены в формате PNG. Признаки экспортированы в файл `features.csv`.
